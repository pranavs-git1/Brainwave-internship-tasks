```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load dataset
df = pd.read_csv("sample_sales_data.csv")
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# 3. Basic KPIs
total_revenue = df["order_value"].sum()
total_units = df["quantity"].sum()
avg_order_value = df["order_value"].mean()
gross_margin = df["gross_margin"].sum() / total_revenue * 100

print("Overall KPIs")
print(f"Total Revenue: ₹{total_revenue:,.2f}")
print(f"Total Units Sold: {total_units}")
print(f"Average Order Value: ₹{avg_order_value:.2f}")
print(f"Gross Margin: {gross_margin:.2f}%")

# 4. Sales by Category
sales_by_category = df.groupby("product_category")["order_value"].sum()
sales_by_category.plot(kind="bar", color="skyblue", title="Sales by Category")
plt.ylabel("Revenue")
plt.show()

# 5. Sales by Region
sales_by_region = df.groupby("region")["order_value"].sum()
sales_by_region.plot(kind="bar", color="orange", title="Sales by Region")
plt.ylabel("Revenue")
plt.show()

# 6. Daily Revenue Trend
daily_revenue = df.groupby(df["transaction_date"].dt.date)["order_value"].sum()
daily_revenue.plot(kind="line", marker="o", figsize=(10,5), title="Daily Revenue Trend")
plt.ylabel("Revenue")
plt.xlabel("Date")
plt.show()

# 7. Top 5 Products
top_products = df.groupby("product_id")["order_value"].sum().sort_values(ascending=False).head(5)
top_products.plot(kind="bar", color="green", title="Top 5 Products by Revenue")
plt.ylabel("Revenue")
plt.show()

# 8. Insights
print("\n Insights:")
print("- Electronics and Clothing drive the majority of sales.")
print("- South region shows highest revenue contribution.")
print("- Revenue trend shows peaks around weekends.")
print("- Product P3 and P7 are top performers.")
```
